```python
def read_dna(file_name):
    with open(file_name,'r') as f:
        dnas = []
        while True:
            dna = f.readline().split('\n')[0]
            if(dna == ''):
                break
            dnas.append(dna)
        f.close()
        return dnas
dnas = []
for file in file_list:
    dnas.append(read_dna(file))
```

```python
def Gc(dna):
    gc = dna.count('G') + dna.count('C')
    l = len(dna)
    return float(gc) / l
gcs  = []
for dna in dnas:
    for d in dna:
        gcs.append(Gc(d))
data_display(gcs)
```

```python
def run(File_name,Chunk_size,Rs):
    name = File_name
    #name = 'sword.enc'
    record = []
    file_name = 'in/' + name
    out_file_name = 'dna.txt'
    out_csv_name = 'out/' + name.split('.')[0] +'.csv'
    re_file_name = 're/' + name
    chunk_size = Chunk_size
    rs_length = Rs
    alp = 2
    f_in, f_size, chunk_num = pre_process(file_name, chunk_size)
    record.append(file_name)
    record.append(out_csv_name)
    record.append(f_size)
    record.append(Rs)
    record.append(rs_length)
    record.append(chunk_size)
    record.append(chunk_num)

    f = DNAFountain(file_in = f_in, 
                        file_size = f_size, 
                        chunk_size = chunk_size, 
                        alpha = alp,
                        rs = rs_length
                        )

    time, out_num = spray(f,out_file_name,out_csv_name)


    g = Glass(chunk_num,out_file_name,rs = rs_length)
    line = decode(0, g, out_file_name)
    if(-1 == line):
        logging.info('decoding failed')
    g.Save(re_file_name)
    
    #re encode use recommended alpha
    alp = (line - chunk_num) / chunk_num * 1.3
    f = DNAFountain(file_in = f_in, 
                        file_size = f_size, 
                        chunk_size = chunk_size, 
                        alpha = alp,
                        rs = rs_length
                        )
    time, out_num = spray(f,out_file_name,out_csv_name)
    
    record.append(alp)
    record.append(out_num / time)
    record.append(out_num)
    record.append(line)
    
    write_csv(record)
    return record
```
